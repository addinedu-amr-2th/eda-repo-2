```python
def chageloc(frame):
    for i in range(len(frame['근무지'])):
        if (frame['근무지'][i].find('경기') > -1):
            frame['근무지'][i] = '경기도'
        elif (frame['근무지'][i].find('서울') > -1):
            frame['근무지'][i] = '서울특별시'
        elif (frame['근무지'][i].find('인천') > -1):
            frame['근무지'][i] = '인천광역시'
        elif (frame['근무지'][i].find('강원') > -1):
            frame['근무지'][i] = '강원도'
        elif (frame['근무지'][i].find('대전') > -1):
            frame['근무지'][i] = '대전광역시'
        elif (frame['근무지'][i].find('충남') > -1):
            frame['근무지'][i] = '충청남도'
        elif (frame['근무지'][i].find('충북') > -1):
            frame['근무지'][i] = '충청북도'
        elif (frame['근무지'][i].find('세종') > -1):
            frame['근무지'][i] = '세종특별자치시'
        elif (frame['근무지'][i].find('전북') > -1):
            frame['근무지'][i] = '전라북도'
        elif (frame['근무지'][i].find('전남') > -1):
            frame['근무지'][i] = '전라남도'
        elif (frame['근무지'][i].find('광주') > -1):
            frame['근무지'][i] = '광주광역시'
        elif (frame['근무지'][i].find('부산') > -1):
            frame['근무지'][i] = '부산광역시'
        elif (frame['근무지'][i].find('경남') > -1):
            frame['근무지'][i] = '경상남도'
        elif (frame['근무지'][i].find('울산') > -1):
            frame['근무지'][i] = '울산광역시'
        elif (frame['근무지'][i].find('경북') > -1):
            frame['근무지'][i] = '경상북도'
        elif (frame['근무지'][i].find('대구') > -1):
            frame['근무지'][i] = '대구광역시'
        elif (frame['근무지'][i].find('미국') > -1):
            frame['근무지'][i] = '해외'
        elif (frame['근무지'][i].find('아시아') > -1):
            frame['근무지'][i] = '해외'
        elif (frame['근무지'][i].find('멕시코') > -1):
            frame['근무지'][i] = '해외'
        elif (frame['근무지'][i].find('전국') > -1):
            frame['근무지'][i] = '서울특별시'
        elif (frame['근무지'][i].find('경력무관') > -1):
            frame['근무지'][i] = '서울특별시'
```

```python
from folium import utilities
from pyppeteer import launch

async def map_to_png(target, m):
    html = m.get_root().render()
    browser = await launch(headless=True)

    page = await browser.newPage()
    with utilities.temp_html_filepath(html) as fname:
        await page.goto('file://{path}'.format(path=fname))

    img_data = await page.screenshot({'path': f'{target}.png', 'fullPage': 'true', })
    await browser.close()

my_map.save('../../data/auto.html')
await map_to_png('자율주행', my_map)
```
